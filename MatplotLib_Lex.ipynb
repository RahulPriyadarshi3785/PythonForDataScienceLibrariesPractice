```python
print('''The subplots method is used to create a common layout for multiple plots. 
'm x n' number of plots can be created using a subplot where m and n represent the number of rows and columns respectively.''')

#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1, 2)    
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()    #adding legend
ax2.legend()
print()
print('y_label of the second plot is overlapping with the first plot')
```

```python
print(' avoid overlapping of y_label of the second plot with the first plot')
print()
#creating sample data for plotting
x = np.arange(1,16)    #1 â€¦ 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#initializing two axes for the subplot
fig, (ax1, ax2) = plt.subplots(1, 2)    
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
fig.tight_layout() # avoid overlapping of y_label of the second plot with the first plot
ax1.legend()    #adding legend
ax2.legend()
```

```python
#importing the data
df = pd.read_csv('C:\\Users\\hp\\Desktop\\Infy\\PythonForDataScienceCodeData\\auto_mpg.csv')
df.info()
```

```python
#replacing the null values in 'horsepower' with its mean value
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)
df.info()
```

```python
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize = [6,6])
#setting axes as [left, bottom, width, height]
ax = fig.add_axes([0,0,1,1])
ax.set_title('Distribution of mileage')
#plotting boxplot
ax.boxplot(df['mpg'])
fig.show()
```

```python
df.info()
```

```python
df['city_mileage'] = df['mpg'] * 0.75
df.head()
```

```python
#Comparing avarage mpg and city mpg using boxplot
mpg_list = [df['mpg'], df['city_mileage']]
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[6,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Distribution of Average MPG vs City MPG')
ax.set_ylabel('Mileage per gallon')
#plotting box plot
ax.boxplot(mpg_list, widths=0.5)
plt.show()
```

```python
print('correlation between horsepower and mileage')
#visualizing correlation between mileage and horsepower
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[10,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
#plotting scatter plot
ax.scatter(df['horsepower'],df['mpg'])
ax.set_title("Scatter plot of horsepower and mileage",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
fig.show()
```

```python
print('correlation between mileage and horsepower based on the origin of the cars')
#finding correlation between mileage and horsepower 
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[12,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Scatter plot of Horsepower and Mileage based on Origin",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
#assigning different colors for each origin
ax.scatter(df['horsepower'],df['mpg'], c=df["origin"].astype('category').cat.codes)
# astype('category') converts it to categorical data
# as cat.codes converts categorical data to numbers
fig.show()
```

```python
#finding correlation between mileage and horsepower 
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[12,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Scatter plot of Horsepower and Mileage based on Origin",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
#assigning different colors for each origin
origin_list = df["origin"].unique()
colors = plt.cm.jet(np.linspace(0, 1, len(origin_list)))
# plotting based on the origin
for origin,color in zip(origin_list,colors):
    x = df[df["origin"] == origin]["horsepower"]
    y = df[df["origin"] == origin]["mpg"]
    plt.scatter(x,y,color = color,label=origin)
plt.legend()
```

```python
#Creating a grouped DataFrame according to model year
grouped_df = df.groupby(['model_year']).count()[['name']]
grouped_df.reset_index(inplace = True)
#plotting the number of cars in each year
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[7,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Number of cars in each year",fontsize=16)
ax.set_ylabel('Number of cars',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
#plotting bar graph
ax.bar(grouped_df['model_year'], grouped_df['name'])
fig.show()
```

```python
print('''used for continuous data. 
Each bar groups the numbers into intervals (bins) and the height of the bar is 
based on the number of values that fall into the corresponding intervals
''')
#creating a histogram for horsepower
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[8,6])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
#plotting histogram
ax.hist(df['horsepower'], bins=20)
ax.set_title('Distribution of Horsepower',fontsize = 17)
ax.set_ylabel('Frequency',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
fig.show()
```

```python
print('Pie chart: ')
pie_df = pd.DataFrame()
pie_df
```

```python
pie_df['Count'] = df['origin'].value_counts()    #getting the count of â€˜originâ€™ and assigning to the df
pie_df
```

```python
pie_df = pie_df.reset_index()    #re-arranging the index
pie_df
```

```python
pie_df.rename(columns={'index':'Country'},inplace=True)    #re-naming the col
pie_df
```

```python
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[8,6])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Origin of the Cars')
#plotting pie chart
ax.pie(pie_df['Count'], labels=pie_df['Country'])
fig.show()
```

```python
print('''A basic Pie chart has been plotted using the ax.pie() function in matplotlib. This plot could also be made visually appealing by customizing other parameters present in the ax.pie() function.

Following are the parameters:

explode       - To get an elevated view for the selected pie. 

colors          -  To customize the colors for the plot.

autopct       -  To add the percentage of the distribution in the pie chart.

shadow       -  To add shadow to the plot.

startangle   -  To change the starting angle of the pie chart.''')
print()
#importing the required packages
import matplotlib.pyplot as plt
fig = plt.figure(figsize=[8,6])
ax = fig.add_axes([0, 0, 1, 1])
explode = [0.05, 0.1, 0]
colors = ['yellowgreen', 'mediumaquamarine', 'khaki' ]
ax.set_title('Origin of the Cars')
ax.pie(pie_df['Count'], labels=pie_df['Country'],
        autopct='%0.1f%%',explode = explode, 
        shadow = True, startangle = 250,
        colors = colors)
fig.show()
```
